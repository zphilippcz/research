```python
import pickle
import os
import re

pattern = r'\*\*"(.+?)"\*\*'
pattern2 = r'- \*\*(.+?)\*\*'
```

```python
for filename in os.listdir("/Users/zphilipp/notebooks/category_synonym/"):
    if filename.endswith('.pkl'): 
        with open('/Users/zphilipp/notebooks/category_synonym/%s' % filename, 'rb') as file:
            loaded_data = pickle.load(file)
            for line in loaded_data:
                
                match = re.search(pattern, line, re.IGNORECASE)
                #print (match.group(1))

                matches = re.findall(pattern2, line)
                categories = []
                #for extracted_text in matches:
                categories.append({match.group(1) : matches})
                print(categories)
                
```
