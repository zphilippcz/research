```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
pd.set_option('display.max_rows', None)
```

```python
def tokenize(text):
    text = [str(text).strip()]
    vectorizer = CountVectorizer()
    try:
        X = vectorizer.fit_transform(text)
    except ValueError as e:
        print ("Stopword only detected <%s>" % text)
        return None
    tokens = vectorizer.get_feature_names_out()
    token_indices = X.toarray().nonzero()[1]
    ordered_tokens = [tokens[i] for i in token_indices]

    return " ".join(map(str.strip, ordered_tokens))
```

```python
df = pd.read_csv("ga.search.query.csv", delimiter = ",")

# clean up the mess
df['query'] = df['query'].apply(tokenize).dropna()
df = df[(df['query'].str.len() > 1) & (df['query'].str.strip() != '')]
df = df[["query", "count"]]
```

```python
df_cleaned = df.groupby('query', as_index=False)['count'].sum().reset_index().sort_values("count", ascending = False)
df_cleaned.head(10)
```
