```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
data = pd.read_csv('q.data.csv', delimiter=";")
#data.query("query == 'oil '")
```

```python
data["query"] = data["query"].str.lower().str.strip()
```

```python
duplicates = data['query'][data['query'].duplicated(keep=False)]
```

```python
duplication_counts = data['query'].value_counts()
data['counts'] = data['query'].map(duplication_counts)
```

```python
df = data.groupby(["query"]).max().sort_values(["counts", "query"], ascending = False)
df.head(100)
```

```python
df = df.reset_index()
df[["query","clicks","conv_rate_fci","fci_conv","fci_ogp","gp_user_fci","gppi_fci","impressions","lci_conv","lci_ogp","search_ctr","search_counts","unique_users"]].to_csv('query.clean.csv', sep=';', index=False)
df.head()
```
